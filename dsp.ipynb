```python
from html.parser import HTMLParser
from bs4 import BeautifulSoup
import requests
from PIL import Image
import io
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import display
```

```python
import concurrent
def request_get(url):
    return requests.get(url, verify=False)

res = {}
with concurrent.futures.ThreadPoolExecutor(max_workers=32) as executor: # optimally defined number of threads
    for x in soup.find_all(class_="item_icon_container"):
        a = x.find("a")
        url_href = a.get("href")
        name = url_href.strip('/')
        if name == "Blank" or name in res:
            continue
        url_img = x.find("img").get("src")
        r1 = executor.submit(request_get, url = f"{URL_BASE}{url_href}")
        r2 = executor.submit(request_get, url = f"{URL_BASE}{url_img}")
        res[name] = (r1,r2)
        print(url_href,url_img)
    executor.shutdown(wait=True)
```

```python
for k,v in res.items():
    print(k)
    code0 = v[0].result().status_code
    code1 = v[1].result().status_code
    if code0!=200:
        print("0:",code0)
    if code1!=200:
        print("1:",code1)
```

```python
for k,v in res.items():
    r1,r2 = v
    res[k] = (r1.result(),r2.result())
```

```python
class DspItem:
    def __init__(self, name, img=None, items_src=None, items_tgt=None):
        self.name = name
        self.img = img
        self.items_src = items_src
        self.items_tgt = items_tgt
    
```

```python
import traceback
items = []
for item_target,v in res.items():
    r1,r2 = res[item_target]
    img=Image.open(io.BytesIO(r2.content))
    try:
        soup1 = BeautifulSoup(r1.content)
        pc_table = soup1.find(class_="wikitable pc_table")
        data = pc_table.find_all('tr')
        header = data[0]
        data = data[1:]
        header_names = [th.text.strip() for th in header.find_all('th')]
        ind = header_names.index('Recipe')
        recipes = [formula.find_all('td')[ind] for formula in data]
        item_source = [x.find('a').get('href').strip('/') 
                       for recipe in recipes
                       for x in recipe.find_all(class_="tt_recipe_item")]
    except Exception as e:
        display(img)
        print(item_target,"parse item_source error")
        print(traceback.format_exc())
        item_source = None
    items.append(
        DspItem(name=item_target,img=img,items_src=item_source)
                )
```

```python
name2item = {}
for item in items:
    name2item[item.name] = item
    item.items_tgt = set()
new_items = []
for item in items:
    if item.items_src is None:
        continue
    for name in item.items_src:
        if name not in name2item:
            item_new = DspItem(name,items_tgt=set())
            new_items.append(item_new)
for item in new_items:
    name2item[item.name] = item
    items.append(item)
for item in items:
    if item.items_src is None:
        continue
    for name_src in item.items_src:
        name2item[name_src].items_tgt.add(item.name)
```

```python
for k,v in name2item.items():
    print(k)
    display(v.img)
```

```python
basic_src = {"Iron_Ingot","Copper_Ingot","Stone_Brick","Glass","Steel",
             "Circuit_Board","Magnetic_Coil","Gear","Plasma_Exciter"}
exclude_tgt = {"Water","Prism","Electromagnetic_Matrix","Electric_Motor","Thruster"}
```

```python
items2show = []
for item in items:
    if item.name in basic_src or item.name in exclude_tgt:
        continue
    if item.items_src is not None and all(x in basic_src for x in item.items_src):
        items2show.append(item)
```

```python
cols = 6
if len(items2show)%cols == 0:
    rows = len(items2show)//cols
else:
    rows = len(items2show)//cols + 1
plt.figure(figsize=(cols*2,rows*2))
for i,item in enumerate(items2show):
    r = i//cols
    c = i%cols
    plt.subplot(rows,cols,i+1)
    plt.imshow(item.img)
    plt.axis("off")
    plt.title(item.name)
plt.show()
```
